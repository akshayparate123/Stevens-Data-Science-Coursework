```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters for the two cases
params_case1 = (4, 98)
params_case2 = (3.5, 102)

# Generate values for the x-axis (theta values)
theta_values = np.linspace(0, 1, 1000)

# Calculate the posterior distributions
posterior_case1 = beta.pdf(theta_values, *params_case1)
posterior_case2 = beta.pdf(theta_values, *params_case2)

# Plot the posterior distributions
plt.plot(theta_values, posterior_case1, label='Case 1: a=b=1')
plt.plot(theta_values, posterior_case2, label='Case 2: a=0.5, b=5')

plt.title('Posterior Distributions')
plt.xlabel('Î¸ (proportion of defective items)')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.show()

# Calculate and compare posterior means
mean_case1 = beta.mean(*params_case1)
mean_case2 = beta.mean(*params_case2)

print(f'Posterior mean for Case 1: {mean_case1}')
print(f'Posterior mean for Case 2: {mean_case2}')
```
