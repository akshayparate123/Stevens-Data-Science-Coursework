```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

# Step 1: Generate Random Data

### • Choose your distribution and specify the parameters.
### • Generate at least 100 data points.
### • Plot the data points on a 2D scatter plot. plot.

```python
# determining the values of the parameters (mean and variance)
mu = 0      #mean
sigma = 1   #variance
n = 100     #total points
alpha = 1   #slope
beta = 2    #intercept

# using the np.random.normal() method for generating 100 random data points.
random_data_points_x = [np.random.normal(mu, sigma) for i in range(0,n)]
# using simple linear regression formula {y = a ∗ x + b + epsilon} for calculating corrosponding y values.
random_data_points_y = [(alpha*i)+beta+np.random.normal(0,0.5) for i in random_data_points_x]
```

```python
plt.scatter(random_data_points_x,random_data_points_y) #Scatter plot
plt.xlabel("x") #x label
plt.ylabel("y") #y label
plt.title("Generated Points") #title
plt.show()
```

# Step 2: Perform Linear Regression Without Neural Networks

### • Fit a linear regression model to the data.
### • Plot the regression line.
### • Comment on the mathematical operations used (e.g., computing the best fit line).

```python
slope = 0    #assigning initial value of slope
sum_of_squared_residuals_list = []    #empty list to store sum_of_squared_residuals
num_epoch = 1000    #number of epochs
slopeList = []    #empty list to store change in slope values
interceptList = []    #empty list to store change in intercept values
for epoch in range (0,num_epoch):
    slope = 0.01+slope    #Change in slope
    intercept = 0    #initial value of intercept
    for counter in range(0,100):
        intercept = 0.1+intercept    #change in intercept
        sum_of_squared_residuals = 0
        for x,y in zip(random_data_points_x,random_data_points_y):    #changing the intercept while keeping the slope constant
            predicted_y = (slope * x) + intercept    #predicting y = slope * x + intercept
            sum_of_squared_residuals = sum_of_squared_residuals+(y-predicted_y)**2    #Adding sum of squared residuals
        sum_of_squared_residuals_list.append(sum_of_squared_residuals)
        slopeList.append(slope)
        interceptList.append(intercept)
```

```python
least_value_index = sum_of_squared_residuals_list.index(min(sum_of_squared_residuals_list)) #Finding the index of minimum sum of squared residual recorded
least_value_index
```

```python
#slope and intercept where the sum of squared residuals was minimum
slopeList[least_value_index],interceptList[least_value_index]
```

```python
#Calculating y using slope and intercept where the sum of squared residuals was minimum
bestFitLine_y = [(slopeList[least_value_index]*i)+interceptList[least_value_index] for i in random_data_points_x]
```

```python
#Plotting the best fit line along with the data points
plt.scatter(random_data_points_x,random_data_points_y,label = "Data points") #Scatter plot
plt.plot(random_data_points_x,bestFitLine_y, color = "red",label = "Best fit line")
plt.xlabel("x") #x label
plt.ylabel("y") #y label
plt.title("Generated Points") #title
plt.legend()
plt.show()
```

### Using sklearn

```python
#Creating linear regression object
lin_reg = LinearRegression()

# Train the model
lin_reg.fit(np.array(random_data_points_x).reshape(-1,1), random_data_points_y)
# Print the parameters
print(f"Intercept: {lin_reg.intercept_}")
print(f"Coefficient: {lin_reg.coef_}")
# Plotting the best fit line and data points
plt.scatter(random_data_points_x, random_data_points_y, color="blue", label="Data points")
plt.plot(random_data_points_x, lin_reg.predict(np.array(random_data_points_x).reshape(-1,1)), color="red", label="Best fit line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Best Fit Line")
plt.show()
```

# Step 3: Perform Linear Regression Using a Neural Network (Single Dense Layer)

### • Build a neural network with 1 dense layer.
### • Train it using a suitable optimizer and loss function (e.g., mean squared error).
### • Plot the predicted line and compare with the linear regression model.
### • Explain why the ‘Dense‘ layer works for this task and comment on the loss function used.

```python
def gradient_descent(a,b,da,db,lr):
    return a-(lr*da),b-(lr*db)
def gradient(x,y_true,y_pred):
    #Calculating Gradient
    gradient_slope = (-2/len(x))*np.sum(x*(np.subtract(y_true,y_pred)))
    gradient_intercept = (-2/len(x))*np.sum(np.subtract(y_true,y_pred))
    return gradient_slope,gradient_intercept
```

```python
num_epoch = 2000
lr = 0.01
w = 0.1 #Weight (slope)
b = 0.2 #Bias (intercept)
loss_list = []
w_list = [];b_list = []
convergence_point = 0; flag = True
for epoch in range(0,num_epoch):
    w_list.append(w)
    b_list.append(b)
    total_loss = 0
    y_pred_list = []
    for x,y in zip(random_data_points_x,random_data_points_y):
        y_pred = w*x + b
        loss = (y - y_pred)**2
        total_loss = total_loss+loss
        y_pred_list.append(y_pred)
    loss_list.append(total_loss/len(random_data_points_x))
    dw,db = gradient(random_data_points_x,random_data_points_y,y_pred_list)
    w,b = gradient_descent(w,b,dw,db,lr)
    if len(loss_list)>2 and (np.abs(loss_list[-1] - loss_list[-2]) < 0.005) and flag:
        convergence_point = epoch
        flag = False
```

```python
min_error_index = loss_list.index(min(loss_list))
```

```python
min_error_index
```

```python
plt.plot(loss_list)
plt.plot(convergence_point, loss_list[convergence_point], marker='o',label = "Convergence Point")
plt.plot(min_error_index, loss_list[min_error_index], marker='o',label = "Minimum loss Point")
plt.title("MSE loss")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
w_list[min_error_index],b_list[min_error_index]
```

```python
bestFitLine_y_1 = [(w_list[min_error_index]*i)+b_list[min_error_index] for i in random_data_points_x]
```

```python
plt.scatter(random_data_points_x,random_data_points_y,label = "Data points") #Scatter plot
plt.plot(random_data_points_x,bestFitLine_y_1, color = "red",label = "Best fit line NN")
plt.xlabel("x") #x label
plt.ylabel("y") #y label
plt.title("NN : 1 Dense Layer ") #title
plt.legend()
plt.show()
```

# Step 4: Experiment with Multiple Layers

### • Build neural networks with 1, 2, 3, and 4 layers.
### • Train each model and plot the predictions.
### • Compare the performance of each model (e.g., using Mean Squared Error)and display the results on a plot.
### • Explain how adding layers changes the model’s performance and complex-ity.

```python
def feed_forward(w1,b1,w2,b2,x,y):
    h1 = w1 @ x.T + b1
    y_pred = w2@h1 + b2
    loss = (y.T - y_pred)
    return loss
def gradient(x,total_loss,h1,w2):
    #Calculating Gradient
    dw2 = ((-2/len(x))*total_loss) @ h1.T #1x64
    db2 = np.sum((-2/len(x))*total_loss, axis=1, keepdims=True) #1x100
    dw1 = (((-2/len(x))*total_loss).T @ w2).T @ x #64x1
    db1 = np.sum(w2.T@ ((-2/len(x))*total_loss), axis=1, keepdims=True) #100x64
    return dw1,db1,dw2,db2
def gradient_descent(w1,b1,w2,b2,dw1,db1,dw2,db2,lr):
    w1 = w1 - (lr*dw1)
    w2 = w2 - (lr*dw2)
    b1 = b1 - (lr*db1)
    b2 = b2 - (lr*db2)
    return w1,b1,w2,b2
def back_propogation(x,loss,h1,w1,b1,w2,b2,dw1,db1,dw2,db2,lr):
    dw1,db1,dw2,db2 = gradient(x,loss,h1,w2)
    w1,b1,w2,b2 = gradient_descent(w1,b1,w2,b2,dw1,db1,dw2,db2,lr)
    return w1,b1,w2,b2
def predict(w1,b1,w2,b2,x,y):
    h1 = w1 @ x.T + b1
    y_pred = w2@h1 + b2
    return y_pred
```

```python
num_epoch = 1000
lr = 0.001
w1 = np.random.randn(64, 1)  # 64 neurons, each connected to 1 input feature
b1 = np.zeros((64, 1))       # Bias for 64 neurons
w2 = np.random.randn(1, 64)  # 1 output neuron connected to 64 hidden neurons
b2 = np.zeros((1, 1))        # Bias for the output neuron
x = np.array(random_data_points_x).reshape(100,1)
y = np.array(random_data_points_y).reshape(100,1)
loss_list = []
w_list = [];b_list = []
convergence_point = 0; flag = True
for epoch in range(0,num_epoch):
    w_list.append([w1,w2])
    b_list.append([b1,b2])
    total_loss = 0
    y_pred_list = []
    loss = feed_forward(w1,b1,w2,b2,x,y)
    loss_list.append(np.mean(loss**2))
    w1,b1,w2,b2 = back_propogation(x,loss,h1,w1,b1,w2,b2,dw1,db1,dw2,db2,lr)
    if len(loss_list)>2 and (np.abs(loss_list[-1] - loss_list[-2]) < 0.005) and flag:
        convergence_point = epoch
        flag = False
```

```python
min_error_index = loss_list.index(min(loss_list))
```

```python
min_error_index
```

```python
plt.plot(loss_list)
plt.plot(convergence_point, loss_list[convergence_point], marker='o',label = "Convergence Point")
plt.plot(min_error_index, loss_list[min_error_index], marker='o',label = "Minimum loss Point")
plt.title("MSE loss")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
bestFitLine_y_2 = predict(w1,b1,w2,b2,x,y)
# predict(w,b,random_data_points_x,random_data_points_y)

```
